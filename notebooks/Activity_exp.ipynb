# Diff Avtivity vs tracking_event

```python
import os
import pymysql
import pandas as pd

conn = pymysql.connect(
    host=os.getenv('HOST'),
    port=int(3306),
    user=os.getenv('USER_DB'),
    passwd=os.getenv('PASSWORD_DB'),
    db=os.getenv('DB'),
    charset='utf8mb4')

diff_df = pd.read_sql_query("""select activity_20220808.resource, tracking_event.resourceID, owner 
                                from activity_20220808
                                inner join tracking_event
                                on activity_20220808.owner = tracking_event.userID  
                                WHERE activity_20220808.owner ='5c643a2c-c517-48b5-9a4e-694da5d5e17b';""",
                             conn)
```

```python
diff_df.drop_duplicates().head()
```

```python
activity_resource = diff_df.resource.unique()
```

```python
event_resource = diff_df.resourceID.unique()
```

```python
unique_activity = []

for x in activity_resource:
    if x not in event_resource:
        unique_activity.append(x)

unique_activity
```

```python
unique_event = []

for y in event_resource:
    if y not in activity_resource:
        unique_event.append(y)
        
unique_event
```

```python
len(diff_df.resource.unique()) - len(diff_df.resourceID.unique())
```

```python
activity_df[activity_df['resource'] == unique_activity[0]] 
```
